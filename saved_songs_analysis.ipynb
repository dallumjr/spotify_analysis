```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates


sp_df = pd.read_csv('spotify_analysis.csv')
sp_df=sp_df[['name', 'added_at', 'danceability', 'energy', 'key', 'mode', 'acousticness','instrumentalness', 'liveness','valence','tempo','time_signature']]
sp_df = sp_df.sort_values(by='added_at')
sp2015_df = sp_df[(sp_df['added_at'] < '2017-01-01') & (sp_df['added_at'] >= '2016-01-01')]
```

```python
sp_df['added_at'] = pd.to_datetime(sp_df['added_at']).dt.strftime('%Y-%m-%d').astype('datetime64[ns]')
```

```python
fig = plt.figure(figsize=(20,10))
ax = sp_df['valence'].plot.hist(bins=10, alpha=0.5, color='green',title='Valence')
ax.set_xlabel('Valence')
ax.set_ylabel('Songs')

fig2 = plt.figure(figsize=(20,10))
ax2 = sp_df['danceability'].plot.hist(bins=10,title='Danceability')
ax2.set_xlabel('Danceability')
ax2.set_ylabel('Songs')

fig3 = plt.figure(figsize=(20,10))
ax3 = sp_df['energy'].plot.hist(bins=10,color='purple', title='Energy')
ax3.set_xlabel('Energy')
ax3.set_ylabel('Songs')

fig4 = plt.figure(figsize=(20,10))
ax4 = sp_df['instrumentalness'].plot.hist(bins=10,color='orange',alpha=0.3, title='Instrumentalness')
ax4.set_xlabel('Instrumentalness')
ax4.set_ylabel('Songs')

fig5 = plt.figure()
ax5 = sp2015_df.plot.scatter(x='added_at',y='valence', color='red',alpha=0.7, title='Valence over Time',figsize=(20,10))
```
